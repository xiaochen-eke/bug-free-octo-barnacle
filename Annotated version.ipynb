```python
#自定义分词器
import re 
text = "Hello, world! This is a test."
# result = re.split(r'(\s),text)')  #按正则表达式拆分字符串，按照空白字符
result = re.split(r'(\s|,|\.|!|\?)', text) #匹配括号里面的的字符如空格逗号等
result = [item for item in result if item.strip() != ""] #去掉多余字符
print(result)
```

```python
preprocessed = re.split(r'(\s|,|\.|!|\?)', text)
preprocessed = [item for item in preprocessed if item.strip() != ""]

#构建词汇表
all_tokens = sorted(set(preprocessed))   #set(preprocessed)是去重，sorted()是排序

vocab = {token:integer for integer, token in enumerate(all_tokens)}#enumerate(all_tokens)是给每个元素自动编号

print(vocab)
```

```python
#第一个分词器
class SimpleTokenizerV1: #定义分词器类
    
    def __init__(self, vocab):
        self.str_to_int = vocab  #将词汇表作为类属性的存储，在encode和decode方法中使用
        self.int_to_str = {i:s for s,i in vocab.items()} #建立反向词典

    def encode(self, text): #将文本转化为id
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)  #正则
        preprocessed = [item.strip() for item in preprocessed if item.strip()] #删除
        ids = [self.str_to_int[token] for token in preprocessed] #转id
        return ids

    def decode(self, ids): #将id转化为文本
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

```python
#第二个分词器
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i: s for s, i in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
        preprocessed = [
            item if item in self.str_to_int  #判断token是否在词汇表中，如果不在就替换为<|unk|>
            else "<|unk|>" for item in preprocessed
        ]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

```python
#BPE分词器 
#BPE=分词，能处理未知单词，基于子句拆分
tokenizer = tiktoken.get_encoding("gpt2") #使用gpt2发BPE词表
text = (
    "Hello, do you like tea? <|endoftext|> In the sunlit terraces "   
    "of someunknownPlace."
)  #<|endoftext|>是GPT的特殊标记

integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print(integers)
```
